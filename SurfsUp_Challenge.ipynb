```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
# print(june_temps)
```

```python
#  3. Convert the June temperatures to a list.
june_temp_list = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
# june_temp_list
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp_df = pd.DataFrame(june_temp_list)
june_temp_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
# print(dec_temps)
```

```python
# 7. Convert the December temperatures to a list.
dec_temp_list = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
# dec_temp_list
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temp_df = pd.DataFrame(dec_temp_list)
dec_temp_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temp_df.describe()
```

## Additional Queries for Getting Weather Data

```python
%matplotlib inline
from matplotlib import style
import matplotlib.pyplot as plt
style.use('bmh')

import seaborn as sns
from scipy import stats
```

```python
# Plot June temps

june_temp_df = pd.DataFrame(june_temp_list, columns = ['tobs'])
june_temp_df.plot.hist(bins=15)
plt.tight_layout()
plt.title('June Temperatures')
plt.xlim([55, 85])
plt.ylim([0,400])
```

```python
# June precipitation
june_temp_prcp = session.query(Measurement.tobs, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
june_prcp_df = pd.DataFrame(june_temp_prcp)
# print(june_prcp_df)
june_prcp_df.describe()
```

```python
# June graph of temp x precipitation with trendline

june_prcp_df = pd.DataFrame(june_temp_prcp, columns=['tobs','prcp'])
#drop NaNs to plot on scatterplot
june_prcp_df = june_prcp_df.dropna()
june_prcp_df.plot.scatter('tobs','prcp')
plt.tight_layout()
plt.title('June Temperature by Precipitation')
plt.xlim([55, 85])
plt.ylim([0, 8])

#trendline
slope, intercept, r_value, p_value, std_err = stats.linregress(june_prcp_df['tobs'],june_prcp_df['prcp'])
sns.regplot(x='tobs',y='prcp', data=june_prcp_df, ci=None, label='y={0:.4f}x+{1:.4f}'.format(slope,intercept)).legend(loc="best")
```

```python
# December precipitation
dec_temp_prcp = session.query(Measurement.tobs, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
dec_prcp_df = pd.DataFrame(dec_temp_prcp)
# print(dec_prcp_df)
dec_prcp_df.describe()
```

```python
# December graph of temp x precipitation with trendline

dec_prcp_df = pd.DataFrame(dec_temp_prcp, columns=['tobs','prcp'])
#drop NaNs to plot on scatterplot
dec_prcp_df = dec_prcp_df.dropna()
dec_prcp_df.plot.scatter('tobs','prcp')
plt.tight_layout()
plt.title('December Temperature by Precipitation')
plt.xlim([55, 85])
plt.ylim([0, 8])

#trendline
slope, intercept, r_value, p_value, std_err = stats.linregress(dec_prcp_df['tobs'], dec_prcp_df['prcp'])
sns.regplot(x='tobs',y='prcp', data=dec_prcp_df, ci=None, label='y={0:.4f}x+{1:.4f}'.format(slope,intercept)).legend(loc="best")
```
